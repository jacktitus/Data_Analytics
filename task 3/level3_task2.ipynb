```python
import pandas as pd
```

```python
df = pd.read_csv("../Dataset .csv")
df.head()
```

Votes analysis
1.Identify the restaurants with the highest and lowest number of votes.

```python
min_votes_restaurants = df.iloc[df['Votes'].idxmin()]
max_votes_restaurants =df.iloc[df['Votes'].idxmax()]

print(f"Restaurant with the lowest number of votes:\n{min_votes_restaurants[['Restaurant Name', 'Votes']]}\n\n")

print(f"Restaurant with maximum number of votes:\n{max_votes_restaurants[['Restaurant Name','Votes']]}")

```

2.Analyze if there is a correlation between the number of votes and the rating of a restaurant.

```python
correlation = df['Votes'].corr(df['Aggregate rating']).round(2)
print(correlation)
```
